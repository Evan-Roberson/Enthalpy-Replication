```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('data.txt',skiprows=1)

Elapsed=np.array(data[:,1]/100000)
Temp=np.array(data[:,2])
Density=np.array(data[:,3])
Press=np.array(data[:,4])
Volume=np.array(data[:,5])
PotEng=np.array(data[:,6])
KinEng=np.array(data[:,7])
TotEng=np.array(data[:,8]/1000)
Enthalpy=np.array(data[:,9]/1000)

data = np.loadtxt('data2.txt',skiprows=1)

Elapsed2=np.array(data[:,1]/100000)
Temp2=np.array(data[:,2])
Density2=np.array(data[:,3])
Press2=np.array(data[:,4])
Volume2=np.array(data[:,5])
PotEng2=np.array(data[:,6])
KinEng2=np.array(data[:,7])
TotEng2=np.array(data[:,8]/1000)
Enthalpy2=np.array(data[:,9]/1000)
```

```python
plt.plot(Elapsed,Temp,'r-')
plt.xlabel('Time,ns')
plt.ylabel('Temperature,K')
plt.show()
plt.plot(Elapsed2,Temp2,'b-')
plt.xlabel('Time,ns')
plt.ylabel('Temperature,K')
plt.show()
```

```python
l=210
k=-325
y=np.polyfit(Temp,Enthalpy,2)
f=np.poly1d(y)
T=np.linspace(Temp[0],Temp[-1],1231)
H=f(T)

dH=np.gradient(H,edge_order=1)
dT=np.gradient(T,edge_order=1)
dHdT=abs(dH/dT)
plt.subplot(2,3,1)
plt.plot(Temp,Enthalpy,'b-')
plt.grid()
plt.axvline(x=l)
plt.axhline(y=k)
plt.xlabel('Temperature,K')
plt.ylabel('Enthalpy, kJ/mol')

plt.subplot(2,3,2)
plt.plot(Temp,H+2,'g-')
plt.grid()
plt.axvline(x=l)
plt.axhline(y=k)

plt.subplot(2,3,3)
plt.plot(Temp,dHdT*100+2,'r-')
plt.axvline(x=l)
plt.grid()
plt.xlabel('Temperature,K')
plt.ylabel('Heat Capacity,Cp J/molK')

y=np.polyfit(Temp2,Enthalpy2,2)
f=np.poly1d(y)
T=np.linspace(Temp2[0],Temp2[-1],1231)
H=f(T)

dH=np.gradient(H,edge_order=1)
dT=np.gradient(T,edge_order=1)
dHdT=abs(dH/dT)
plt.subplot(2,3,4)
plt.plot(Temp2,Enthalpy2,'b-')
plt.grid()
plt.xlabel('Temperature,K')
plt.ylabel('Enthalpy, kJ/mol')

plt.subplot(2,3,5)
plt.plot(Temp2,H,'g-')
plt.grid()
plt.axvline(x=l)

plt.subplot(2,3,6)
plt.plot(Temp2,dHdT*100,'r-')
plt.axvline(x=l)
plt.grid()
plt.xlabel('Temperature,K')
plt.ylabel('Heat Capacity,Cp J/molK')
plt.subplots_adjust(left=1,
                    bottom=0.5,
                    right=2.25,
                    top=2,
                    wspace=.7,
                    hspace=.3)
plt.show()
```
